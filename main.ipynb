```python
# Load required packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import joblib
import glob 
import os
```

```python
# Load the data
csv_dir = 'reno9pro'
csv_files = glob.glob(os.path.join(csv_dir, '*.csv'))

# Define the directory to save the model files
model_dir = 'generated_models'
os.makedirs(model_dir, exist_ok=True)  

# Print the files names
for f in csv_files:
    print(f)
```

```python
# Process the data
for f in csv_files:
    # Load the data
    df = pd.read_csv(f)
    
    # Create a new column for time in seconds
    df['Time'] = pd.to_timedelta(df['Time'] + ':00')
    df['Seconds'] = df['Time'].dt.total_seconds().astype(int) 
        
    # Create a scatter plot
    plt.figure(figsize=(10,6))
    sns.scatterplot(x='Seconds', y='Heart Rate', data=df)
    plt.title(f'Scatter Plot for {os.path.basename(f)}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Heart Rate')
    plt.grid(True)
    plt.show()
    
    # Split the data into training and testing data
    X = df['Seconds'].values.reshape(-1, 1)
    y = df['Heart Rate'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    
    # Train the model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Plot the regression line
    plt.figure(figsize=(10,6))
    plt.scatter(X, y, label='Data Points')
    plt.plot(X_test, y_pred, color='red', label='Regression Line')
    plt.title(f'Regression Line for {os.path.basename(f)}')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Heart Rate')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Evaluate the model
    print(f'Model Evaluation for {os.path.basename(f)}:')
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('R2 Score:', metrics.r2_score(y_test, y_pred))
    print('\n')
    
    # Basic statistics for heart rate
    print(f'Basic Statistics for Heart Rate in {os.path.basename(f)}:')
    print('Minimum:', np.min(y))
    print('Maximum:', np.max(y))
    print('Mean:', np.mean(y))
    print('Standard Deviation:', np.std(y))
    print('\n')
    
    # Count-related data
    print(f'Count Related Data for {os.path.basename(f)}:')
    print('Total Data Points:', len(df))
    print('Unique Heart Rate Values:', df['Heart Rate'].nunique())
    print('Percentage of Unique Heart Rate Values:', 
          df['Heart Rate'].nunique() / len(df) * 100)
    print('\n')
    
    # Save the model
    model_name = os.path.basename(f).split('.')[0] + '.pkl'
    model_path = os.path.join(model_dir, model_name)
    joblib.dump(model, model_path)
    print(f'Model saved as: {model_path}\n')
```
